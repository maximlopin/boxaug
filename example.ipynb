```python
from boxaug.transforms import Compose, Affine, Resize, Flip, Color
from boxaug.utils import to_pil
from boxaug.datasets import LabelImgDataset
```

```python
tfm = Compose(
    Resize(128, 128),
    Color(0.3, 0.3, 0.3),
    Affine(deg=0, shear=(0.3, 0.3), resampling='constant'),
    Flip(axis=0),
    Flip(axis=1)
)
```

```python
ds = LabelImgDataset('./data', tfm)
```

```python
for _ in range(10):
    image, bboxes, labels = ds[0]
    display(to_pil(image, bboxes))
```
